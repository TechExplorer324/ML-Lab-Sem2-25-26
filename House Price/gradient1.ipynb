```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (12.0, 9.0)
data = pd.read_csv('gradientData.csv')

X = data.iloc[:, 0]
Y = data.iloc[:, 1]

plt.scatter(X, Y)
plt.show()
```

```python


lr = 0.0001 # The learning Rate
iterations = 20 # The number of iterations to perform gradient descent

```

```python
costs = []

n = float(len(X)) # Number of elements in X

m = 0
b = 0

for i in range(iterations):
    y_pred = m*X + b
    
    cost = (1/n) * sum((Y - y_pred)**2) #MSE
    costs.append(cost)

    dm = (-2/n)*(sum(X*(Y - y_pred)))
    db = (-2/n)*(sum(Y - y_pred))

    m = m-lr*dm
    b = b - lr*db

y = m*X+b

plt.scatter(X, Y)
plt.plot(X, y, color='red')
plt.show()

print(f"Final parameters: m = {m:.2f}, b = {b:.2f}")
print("Change of costs ", costs)
print(f"Final cost: {costs[-1]:.4f}")
    
```

```python
plt.plot(costs)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost vs Iterations')
plt.show()
```
